# Color Compression

The idea is to run k-means algorithm on the color scale of [0, 255] then for each pixel replace the current tuple of RGB values with their representative

```python
import numpy as np
from PIL import Image
import random
import matplotlib.pyplot as plt
import time
import math
```

## Load the image

```python
#Get the image
im = Image.open('bricks.jpeg') # Can be many different formats.
pix = im.load()
print('The size of the image is', im.size)
height = im.size[0]
width = im.size[1]
display(im)
```

## Define necessary functions and constants

```python
iter = 100
INF = 1e9+10
N = 256
color_range = 256
cluster = [-1]*N
Jclust = []
pool = []

for i in range(color_range):
    pool.append(i)

def dist(a, b):
    return abs(a-b)

def centroids_generator(k):
    _centroid = random.sample(range(color_range), k)
    return _centroid
                        
                        
def kmeans(k):
    start_time = time.time()
    Jclust.clear()
    centroid = centroids_generator(k)

    for i in range(iter):
        cost = 0
        for item in range(color_range):
            mn = INF
            label = -1
            for (ind, cent) in enumerate(centroid): 
                if dist(item, cent)<mn:
                    mn = dist(item, cent)
                    label = ind
            cluster[item] = label
            cost+=mn
        Jclust.append(cost/N)
        cnt = [0]*k
        sum = [0]*k
        for key in range(color_range):
            cnt[cluster[key]]+=1
            sum[cluster[key]]+=key

        for j in range(k):
            centroid[j] = sum[j]/cnt[j]
            
    print("--- %s seconds ---" % (time.time() - start_time))
    return centroid
            
#show Jclust
def show_Jclust():
    plt.plot(Jclust)
    plt.show()
```

## Run the algorithm

```python
centroid = kmeans(2)
print(centroid)
show_Jclust()
```

## Replacing the original color with it's representative color

```python
im = Image.open('bricks.jpeg') # Can be many different formats.
pix = im.load()

for i in range(color_range):
    pool[i] = centroid[cluster[i]]
    
for i in range(height):
    for j in range(width):
        pix[i, j] = (int(pool[pix[i, j][0]]), int(pool[pix[i, j][1]]), int(pool[pix[i, j][2]]))

display(im)
```
